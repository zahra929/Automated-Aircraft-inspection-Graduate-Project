```python
IMG_WIDTH = 128
IMG_HEIGHT = 128
IMG_CHANNELS = 3 
```

```python
TRAIN_PATH='/content/drive/My Drive/dataset0/train'

```

```python
X_train = np.zeros ((53 ,IMG_HEIGHT,IMG_WIDTH , IMG_CHANNELS), dtype=np.uint8)
Y_train = np.zeros(( 53 ,IMG_HEIGHT,IMG_WIDTH, 1 ),dtype=np.bool)

print ('resizing training images and masks')
for n in range (1 , 53):
     path = TRAIN_PATH 
     for img_file in next(os.walk(path +'/image/'))[2]:
       img = imread(path + '/image/'+img_file)[:,:,:IMG_CHANNELS]
       img= resize (img, (IMG_HEIGHT,IMG_WIDTH), mode='constant', preserve_range=True)
     X_train[n] = img

     for mask_file in next(os.walk(path +'/mask/'))[2]:
       mask = imread(path + '/mask/'+ mask_file)
       mask = resize (mask, (IMG_HEIGHT,IMG_WIDTH,1), mode='constant', preserve_range=True)
     Y_train[n] = mask 
    
```

Techniques d'augmentation : rotation, flipping, clipping 

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from skimage import io

image_datagen = ImageDataGenerator(featurewise_center=True,  
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        brightness_range=None,
        rescale=1./255,
        shear_range=0.1,
        zoom_range=0.3,
        horizontal_flip=True,
        fill_mode='nearest')

mask_datagen = ImageDataGenerator(featurewise_center=True,       
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        brightness_range=None,
        rescale=1./255,
        shear_range=0.1,
        zoom_range=0.3,
        horizontal_flip=True,
        fill_mode='nearest')

seed = 0
l=-4
for i in range (1,53):
   chemin= '/content/drive/My Drive/dents/train/'+ str(i) +'/images/'+ str(i) + '.jpg'
   cheminn='/content/drive/My Drive/dents/train/'+ str(i) +'/masks/'+ str(i) + '.tiff'
   img = load_img(chemin)  
   mask = load_img(cheminn)
   
   x = img_to_array(img)  
   x = x.reshape((1,) + x.shape)  

   xx = img_to_array(mask)  
   xx = xx.reshape((1,) + xx.shape)  

   image_datagen.fit(x, augment=True, seed=seed)
   mask_datagen.fit(xx, augment=True, seed=seed)
   l+= 4
   j=1
   for j in range (1,6):
    seed +=6

    for chemin in image_datagen.flow(
          x, seed=seed , batch_size=1,
          save_to_dir='/content/drive/My Drive/Dataset3_augm/train/'+ str(j+i+52+l-1) +'/images',
          save_prefix= str(j+i+52+l-1), 
          save_format='jpg'):
       break

    for cheminn in mask_datagen.flow(
          xx,seed=seed , batch_size=1,
          save_to_dir='/content/drive/My Drive/Dataset3_augm/train/'+ str(j+i+52+l-1) +'/masks',
          save_prefix= str(j+i+52+l-1), 
          save_format='tiff'):
       break    
```
